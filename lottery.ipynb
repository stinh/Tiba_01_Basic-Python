樂透中獎機率程式

```python
import random
```

直覺式寫法

```python
# 七個數字的樂透券

# 設置頭獎號碼
prize = set()
while len(prize) < 7:
  n = random.randint(1, 48)
  prize.add(n)
print("頭獎的彩券：",prize)

times = 0
while True:
  # 買彩券
  lottery = set()
  while len(lottery) < 7:
    n = random.randint(1, 48)
    lottery.add(n)
  print("我買的彩券：",lottery)
  times = times + 1

  # 兌獎
  total = 0
  for n in lottery:
    if n in prize:
      print("中了",n, "這個號碼")
      total = total + 1
  print("中了",total,"個數字")

  # 中4個數字(算中獎)，就停止
  if total >= 4:
    break

print("總共買了", times, "張才中")
```

修改寫法

```python
def generate_ticket():
  ticket = set()
  while len(ticket) < 7:
    n = random.randint(1, 48)
    ticket.add(n)
  return ticket()


prize = generate_ticket()
print("頭獎的彩券：",prize)

times = 0
while True:
  # 買彩券
  lottery = generate_ticket()
  print("我買的彩券：",lottery)
  times = times + 1

  # 兌獎
  total = 0
  for n in lottery:
    if n in prize:
      print("中了",n, "這個號碼")
      total = total + 1
  print("中了",total,"個數字")

  # 中4個數字(算中獎)，就停止
  if total >= 4:
    break

print("總共買了", times, "張才中")
```
